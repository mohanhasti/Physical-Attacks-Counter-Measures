```python
tvla_0 = np.load('C:/Users/coe_user/Desktop/PhyAttcks/hw3/data/tvla_traces_hw3/tvla_0traces_int16.npy')
tvla_1 = np.load('C:/Users/coe_user/Desktop/PhyAttcks/hw3/data/tvla_traces_hw3/tvla_1traces_int16.npy')
```

```python
Na = len(tvla_0)
Nb = len(tvla_1)
Xa= np.mean(tvla_0, axis = 0)
Xb = np.mean(tvla_1, axis = 0)

Sa = np.std(tvla_0, axis = 0, where = (tvla_0!=0))
Sb = np.std(tvla_1, axis = 0 ,where = (tvla_1!=0))

numerator = np.subtract(Xa,Xb)
denom_1 = np.divide(np.square(Sa),Na)
denom_2 = np.divide(np.square(Sb),Nb)
denominator = np.sqrt(np.add(denom_1,denom_2))
denominator[denominator == 0] = np.inf
# np.where(denominator == 0 , np.inf, denominator)
t_stat = np.divide(numerator,denominator)
```

```python
import matplotlib.pyplot as plt 
_, ax = plt.subplots(figsize=(10, 3))
x = np.array(range(len(t_stat)))
plt.axhline(y = 4.5, color = 'r', linestyle = '-')
ax.set_title("T-Statistic")
ax.set_xlabel("Traces")
ax.set_ylabel("T-stat values")
ax.plot(x, t_stat, c= 'blue')
plt.show()
```
